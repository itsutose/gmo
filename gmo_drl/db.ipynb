```python
from sqlalchemy import create_engine, MetaData, Table, inspect
import pandas as pd

def db2pd(file_path):
    # Create a connection to the SQLite database
    engine = create_engine(f'sqlite:///{file_path}')
    # engine = create_engine('sqlite://///workspace/gmo_data/ratest_rate/2023-08-20_ratest_rate.db')

    inspector = inspect(engine)
    table_names = inspector.get_table_names()[0]

    # Reflect the trading_history table
    metadata = MetaData()
    trading_history = Table(table_names, metadata, autoload=True, autoload_with=engine)

    # Query all rows of the table
    results = engine.execute(trading_history.select()).fetchall()

    # Convert results to a list of dictionaries for easy viewing
    rows = [dict(row) for row in results]

    # Convert the list of dictionaries to a pandas DataFrame
    df_trading_history = pd.DataFrame(rows).drop('id',axis=1)

    return df_trading_history

# db2pd('/workspace/gmo_data/trading_hist/2023-08-19_trading_hist.db')
# db2pd('/workspace/gmo_data/ratest_rate/2023-08-20_ratest_rate.db')
```

```python
from sqlalchemy import create_engine, MetaData, Table
import pandas as pd

# Create a connection to the SQLite database
engine = create_engine('sqlite:////workspace/gmo_data/board/2023-08-17_board.db')

# Reflect the asks and bids tables
metadata = MetaData()
asks_table = Table('asks', metadata, autoload=True, autoload_with=engine)
bids_table = Table('bids', metadata, autoload=True, autoload_with=engine)

# Query data from the asks table
asks_data = engine.execute(asks_table.select()).fetchall()

# Query data from the bids table
bids_data = engine.execute(bids_table.select()).fetchall()

# Convert results to a list of dictionaries for easy viewing (optional)
asks_rows = [dict(row) for row in asks_data]
bids_rows = [dict(row) for row in bids_data]
```
